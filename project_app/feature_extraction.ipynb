```python
import pandas as pd
import pickle
import numpy as np
```

```python
# all data taken here
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df
# we want this features
# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
# lets check for null values
df.isnull().sum()
```

city have lots of missing values

```python
# here looking at venue we are able to see city names
df[df['city'].isnull()]['venue'].value_counts()
```

```python
# this city names from venue are extracted and fill thus null values
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
# column created
df['city'] = cities
```

```python
# now there is no null values
df.isnull().sum()
```

```python
# venue is not required to us so we will drop that column
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
# now we will take only thus values where at least 5 matches are played means 600 balls are played
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
# filter condition
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
# to carry out current score
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
# thus code is to get over and ball_no columns
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python

df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
# balls_left column is created
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
# to wickets left from player_dismissed column
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
# to get current run rate crr
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# to get last five overs run
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
# column of last_five created
df['last_five'] = last_five
```

```python
# to get final score 
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
# now this is our final_df
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.isnull().sum()
```

```python
# missing values we will remove
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
# to avoid here we do shuffling
final_df = final_df.sample(final_df.shape[0])
```
